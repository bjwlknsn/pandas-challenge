```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
school_file = "../Resources/schools_complete.csv"
student_file = "../Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_df = pd.read_csv(school_file)
student_df = pd.read_csv(student_file)

# Combine the data into a single dataset.  
merge_df = pd.merge(student_df, school_df, how="left", on=["school_name", "school_name"])

#Check merged dataframe
merge_df.head()
```

## District Summary

```python
# Calculate the total number of schools
total_school = len(merge_df["School ID"].unique())

# Calculate the total number of students
total_student = len(merge_df["Student ID"].unique())

# Calculate the total budget
total_budget = school_df["budget"].sum()

# Calculate the average math score
math_avg = merge_df["math_score"].mean()

# Calculate the average reading score
reading_avg = merge_df["reading_score"].mean()

# Calculate the percentage of students with a passing math score (70 or greater)
pass_math_percent = merge_df.loc[merge_df["math_score"]>=70]["math_score"].count()/total_student

# Calculate the percentage of students with a passing reading score (70 or greater)
pass_reading_percent = merge_df.loc[merge_df["reading_score"]>=70]["reading_score"].count()/total_student

# Calculate the percentage of students who passed math and reading (% Overall Passing)
overall_pass_percent = merge_df[(merge_df["math_score"]>=70) & (merge_df["reading_score"]>=70)]["Student ID"].count()/total_student

# Create a dataframe to hold the above results
district_summary_df =  pd.DataFrame({"Total Schools":[total_school],
                                 "Total Students":[total_student],
                                 "Total Budget":[total_budget],
                                 "Average Math Score":[math_avg],
                                 "Average Reading Score":[reading_avg],
                                 "% Passing Math":[pass_math_percent],
                                 "% Passing Reading":[pass_reading_percent],
                                 "% Overall Passing":[overall_pass_percent],
                                 })

# Optional: give the displayed data cleaner formatting
district_summary_df['Total Students'] = district_summary_df['Total Students'].map("{:,}".format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map("${:,.2f}".format)
district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map("{:,.2f}".format)
district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map("{:,.2f}".format)
district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map("{:.1%}".format)
district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map("{:.1%}".format)
district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map("{:.1%}".format)


district_summary_df
```

## School Summary

```python
# School Name
school_name = merge_df.set_index("school_name").groupby(["school_name"])

# School Type
school_types = school_df.set_index("school_name")["type"]

# Total Students
student_total = school_name["Student ID"].count()

# Total School Budget
budget_total = school_df.set_index("school_name")["budget"]

# Per Student Budget
per_student_budget = budget_total/student_total

# Average Math Score
avg_math_score = school_name["math_score"].mean()

# Average Reading Score
avg_reading_score = school_name["reading_score"].mean()

# % Passing Math
percent_pass_math = merge_df.loc[merge_df["math_score"] >= 70].groupby("school_name")["Student ID"].count()/student_total

# % Passing Reading
percent_pass_reading = merge_df.loc[merge_df["reading_score"] >= 70].groupby("school_name")["Student ID"].count()/student_total

# % Overall Passing (The percentage of students that passed math and reading.)
percent_overall_pass = merge_df[(merge_df["math_score"]>=70) & (merge_df["reading_score"]>=70)].groupby("school_name")["Student ID"].count()/student_total

# Create a dataframe to hold the above results
school_summary_df = pd.DataFrame({
                               "School Type" : school_types,
                               "Total Students" : student_total,
                               "Total School Budget" : budget_total,
                               "Per Student Budget" : per_student_budget,
                               "Average Math Score" : avg_math_score,
                               "Average Reading Score" : avg_reading_score,
                               "% Passing Math": percent_pass_math,
                               "% Passing Reading" : percent_pass_reading,
                               "% Overall Passing" : percent_overall_pass
    
                            })

school_summary_df
```

## Top Performing Schools (By % Overall Passing)

```python
#Sort and display the top five performing schools by % overall passing.
top_schools = school_summary_df.sort_values(["% Overall Passing"], ascending=[False])
top_schools.head(5)
```

## Bottom Performing Schools (By % Overall Passing)

```python
#Sort and display the five worst-performing schools by % overall passing.
bottom_schools = top_schools.tail(5)
bottom_schools.sort_values(["% Overall Passing"], ascending=[True])
```

## Math Scores by Grade

```python
# Create a series that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.
math_9th = merge_df.loc[merge_df["grade"] == "9th"].groupby("school_name")["math_score"].mean()
math_10th = merge_df.loc[merge_df["grade"] == "10th"].groupby("school_name")["math_score"].mean()
math_11th = merge_df.loc[merge_df["grade"] == "11th"].groupby("school_name")["math_score"].mean()
math_12th = merge_df.loc[merge_df["grade"] == "12th"].groupby("school_name")["math_score"].mean()


# Combine the series into a dataframe
math_grade_df = pd.DataFrame({"9th": math_9th,
                            "10th": math_10th,
                            "11th": math_11th,
                            "12th": math_12th,
    
                            })

# Optional: give the displayed data cleaner formatting
math_grade_df["9th"] = math_grade_df["9th"].map("{:.1f}".format)
math_grade_df["10th"] = math_grade_df["10th"].map("{:.1f}".format)
math_grade_df["11th"] = math_grade_df["11th"].map("{:.1f}".format)
math_grade_df["12th"] = math_grade_df["12th"].map("{:.1f}".format)

math_grade_df
```

## Reading Score by Grade

## Scores by School Spending

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
bins = [0, 584.99, 629.99, 644.99, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels = group_names)
spending_range = school_summary_df.groupby(["Spending Ranges (Per Student)"])

# Average Math Score
mean_math_score = spending_range["Average Math Score"].mean()

# Average Reading Score
mean_reading_score = spending_range["Average Reading Score"].mean()

# % Passing Math
mean_percent_math = spending_range["% Passing Math"].mean()*100

# % Passing Reading
mean_percent_reading = spending_range["% Passing Reading"].mean()*100

# Overall Passing Rate (Average of the above two)
overall_percent_pass = spending_range["% Overall Passing"].mean()*100

#create a df to store Scores by School Spending
school_spending_df = pd.DataFrame({"Average Math Score":mean_math_score,
                                          "Average Reading Score":mean_reading_score,
                                          "% Passing Math":mean_percent_math,
                                          "% Passing Reading":mean_percent_reading,
                                          "% Overall Passing Rate":overall_percent_pass
                                         })

school_spending_df["Average Math Score"] = school_spending_df["Average Math Score"].map("{:.2f}".format)
school_spending_df["Average Reading Score"] = school_spending_df["Average Reading Score"].map("{:.2f}".format)
school_spending_df["% Passing Math"] = school_spending_df["% Passing Math"].map("{:.2f}".format)
school_spending_df["% Passing Reading"] = school_spending_df["% Passing Reading"].map("{:.2f}".format)
school_spending_df["% Overall Passing Rate"] = school_spending_df["% Overall Passing Rate"].map("{:.2f}".format)

school_spending_df
```

## Scores by School Size
